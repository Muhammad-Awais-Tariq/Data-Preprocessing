## What is Feature engineering ?

Its the process of using the domain knowledge to extract the new features from the raw data and these can be used to improve the performance of ml algorithm

There is no extact steps of the features engineering its unique to the person its not generalized

## Types of feature engineering

1. Feature transformation

If we have one coloumn and we are transforming it into another form all these fall under the feature transformation it includes

- Missing values imputation
- Handling categorical features
- Outlier detection
- Feature scaling

2. Feature construction

Sometime the features are not present in the data but we deduce that if those are present they will improve the performance so we construct em

3. Feature selection

It mean that instead of passing all the features to the algorithm we select specific features to pass

4. Feature Extraction

Programmatically using a algorithm extracting new features from the given features without the manual methood

## Coloumn Transformers

Each coloumn has different problems like age has missing values some has ordinal encoding and some have the ont hot encoding we will have to do this sepratly

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder


transformer = ColumnTransformer(transformers = [
    ('tnf1' , SimpleImputer() , ["It will have coloumn names"]  ), #these are the tuples or transformation we are going to apply
    ('tnf2' , OrdinalEncoder(categories=[["we will pass categories here"]]) , ["coloumn"]),
    ('tnf3' , OneHotEncoder(sparse=False, drop="first"))
] , remainder='passthrough')

transformer.fit_transform() # x train
transformer.transform() # x test
```

The transformers have all the coloumn the remainder has the option for the coloumns in which we will apply no transformation we can leave them like using remainder = passthrough or we can drop em using drop

The transfomers take in tuples which are the number of transfomartions that we want to apply each tuple take three arguments first is the name of the tuple can by any second the type of transformation we need to apply and last all the coloumns where we want to apply it

## Piepelines

Its a mechanisim to chain multiple steps together so the output of the one process can be used as the input of another process

They make it easy to apply the same preprocessing to train and test

If we do not use the pipelines the production code will be hard as we have to manually apply all the encodings etc this will break the code so on easy step for it is to use the pipelines like

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder , MinMaxScaler
from sklearn.pipeline import Pipeline , make_pipeline
from sklearn.feature_selection import SelectKBest,chi2
from sklearn.linear_model import LinearRegression
```

Now we will plan on what to do with the data

For exmaple we have missing data in two coloumns like age and embarked first we will fill the missing values using coloumn transformer in the pipeline and then will send the data to the next step to one hot encode the data and then we will do the scaling and then in last we will do the feature selection it will select the top coloumns not required for all but used

```python
trf1 = ColumnTransformer([
    ("impute_age" , SimpleImputer() , [2]), # instead of using the coloumn names its a good strategy to use the indexes
    ("impute_embarked" , SimpleImputer(strategy="most_frequent") , [6])
],remainder="passthrough")
```

We use the indexes instead of coloumn name when using the transfomer in the pipeline beacuse it convert the df into the numpy array which can break the steps if the next transfomer is accepting a coloumn name thats why we use the indexes

```python
trf2 = ColumnTransformer([
    ("ohe_sex_embarked" , OneHotEncoder(sparse=False , handle_unknown="ignore") , [1,6]) # multiple coloumns
],remainder="passthrough")
```

We can use the standard scaler normally but since we are using the feature selection thats why we are doing the min max scaler 

```python
trf3 = ColumnTransformer([
    ('scale' , MinMaxScaler() , slice(0,10)) # slice will give all the coloumns in the given range
])
```

```python
trf4 = SelectKBest(score_func=chi2 ,k=8) # we can skip it its feature selection
```

```python
trf5 = LinearRegression() # final model
```

```python
pipe = Pipeline([
    ("transformation1" , trf1),
    ("transformation2" , trf2),
    ("transformation3" , trf3),
    ("transformation4" , trf4),
    ("transformation5" , trf5)
])
```

We are passing all the steps of the transformation in the pipeline

Instead of the pipline we can simply use the make pipline its simpler the pipeline as in it we do not have to give name to each transformation so overall its simple same way we can use the make_coloumn as well

```python
pipe = make_pipeline(trf1 , trf2 , trf3 , trf4 , trf5)
pipe.fit() # xtrain y train
```

If we didnt have to last step meaning the model training then instead of fit we would have called the pipe.fit_transform() to preprocess the data

```python
# we can display the pipeline using
from sklearn import set_config

set_config(display="diagram")
```

```python
# we can explore the piplines using
pipe.named_steps['trf1'].transformers_[0] #using these steps and different attributes from the sklearn website to find different function
```

If we want to to the test like passing in the number of values to check the performance of our model we can do it using the dridsearch like

```python
# parameter to changes

params = {
    'trf5_max_depth' : [1,2,3,4,5,None] # the name of the transformer step as well as paramter step

}

from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(pipe , params , cv=5 , scoring="accuracy")

grid.best_params_ # to tell the best parameter
```

#### Discretization

Its process of transforming continuous variables into discrete varaibles by createing a set of contiguous intervals that span the range of varaible values

In simple words if we have coloumns like age and it has values from 1 to 100 we can create bins like one from 1 to 10 then from 10 to 20 so this is discretization in simple words

#### Benefits of discretization

- To handle outlier
- to improve the value spread

##### Types of discretization

1. Unsupervised binning
2. Supervised binning
3. Custom binning

Each of type is further classfied as

1. Unsupervised binning

- Equal width binning (also called uniform)
- Equal frequency binning (also called quantile)
- kmeans binning

2. Supervised binning

- Decesion tree binning

## Handling missing data

This approach is called Complete case analysis formally its the approach of cases consists in discarding observation where values in any of the varaibles are missing

If we have missing values in the coloumns we have only two options these are
1. Remove the row completly that has missing coloumn (but not prefered as we are removing usefull data as well)
2. Impute the value (it is done using the simple imputer class)
- Univariate (depending on numerical or categorical data it has diff techniques)
- multivariate(KNN inmuter or iterator imputator)

### Assumption for cca

The assumption of cca is that missing values at random these values are missing random not following the pattern and we remove the rows only when the values are missing randomly

### Pros / Cons of CCA

#### Pros

- Easy to implment no data manipulation required
- It preserve the distribution if the values are missing at random

#### Cons

- It can exclude a large fraction of original data
- If the data is not missing at random then we remove the important observation for analysis
- When the model is used are in the production the model will not know how to handle missing data

#### When to use CCA

- We can use it when the data is missing at random
- if the data missing is larger then 5% we do not use it

## Handling missing numerical values

We can fill the numerical data using the two ways like
- univariant imputation (in it if the value is missing in a row we use that coloumn to find the value)
- multivarant imputation (if in order to fill it we are using multiple coloumns)

#### Mean/median imputation

If we have a coloumn like age and it has missing values we can fill those values with the mean or median of the entire coloumn 

##### When to use mean or median ?

Use mean when the data is normally distributed and if the distribution is skewed use the mode

- This technique is simple so can easily be implmented in production code

Disadvante

The distvantage of it is that it changes the distribution of the data and add the outliers and effect the relation of one coloumn with the other coloumn

#### When to use ?

If the data is missing completly at random and the data missing is less then 5%

Code

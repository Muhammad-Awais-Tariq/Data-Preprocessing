## What is Feature engineering ?

Its the process of using the domain knowledge to extract the new features from the raw data and these can be used to improve the performance of ml algorithm

There is no extact steps of the features engineering its unique to the person its not generalized

## Types of feature engineering

1. Feature transformation

If we have one coloumn and we are transforming it into another form all these fall under the feature transformation it includes

- Missing values imputation
- Handling categorical features
- Outlier detection
- Feature scaling

2. Feature construction

Sometime the features are not present in the data but we deduce that if those are present they will improve the performance so we construct em

3. Feature selection

It mean that instead of passing all the features to the algorithm we select specific features to pass

4. Feature Extraction

Programmatically using a algorithm extracting new features from the given features without the manual methood

# Feature Scaling

Its usually the last step in the ml pipeline onces we have done the data cleaning missing values categorical values etc then we do feature scaling

Formally feature scaling is the technique use to standardize the independent features present in the data in a fixed range

### Types of Feature Scaling

It has basically two types

1. Stndardization
2. Normalization

### Stndardization

Its also called z score normalization

![](Images/standardization_formula.png)

## Encoding Categorical data

Categorical data can be off two types

1. Nominal
its the type in which the categories does not have any relation or order with one another
like state of a country there is no order to the state of that country like that state value is higher then that so its norminal

2. Ordinal its the type in which the categories have relation or order like marks in the high school like distinction first class second class and one more example can be like review excellent good etc these are all ordinal

### How to encode categorical data

The categories are mostly in the numeric form but the ml model expects the data to be in the form of the numbers so we can do it using the different techniques which are

- Ordinal Encoding
- One hot encoding

### Label encoding

Its same as the oridinal encoding with a small differencce

In the data we have the input columns and the output coloumn if the input coloumns has a coloumn that has oridinal categorical data we will use ordinal encoding there but  if the output label y is categorical like classification problem so for it we will not use the oridinal encoding we will use the label encoding for it it does the same job as the oridinal encoding but its designed for the output label

## Oridinal encoding

If we have the data like the education which is the categorical data and we know that post graduation is higher then under graduation meaning that its ordinal categorical data so we will use ordinal encoding

In it we will specify that give the post gradution highest value and under graduation lower then high school lowset so the ordinal encoding will automaticall assign the required values

![](Images/ordinal_encoding.png)

```python
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder(categories=[['School' , 'UG' , 'PG']]) # we are specifiying the order for encoding

oe.fit() #train
x_train = oe.transform() #train
```
